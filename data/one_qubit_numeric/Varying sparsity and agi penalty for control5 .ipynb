```python
import dill
from convex import *
with open('pickled_controls5.pkl', 'rb') as data:
    data = dill.load(data)
```

```python
derivs = all_derivs(data.controlset, data.target_operator, data.control_hamiltonians, data.ambient_hamiltonian,
                    data.dt, 1)
```

```python
sample_values = [0, .0001, .001, .01, .1]
all_the_weights_sparsity = []
for param in sample_values:
    weights_0 = optimal_weights(derivs[:1], l2_constraint=False, sparsity=True, sparsity_param=param)
    weights = optimal_weights(derivs, l2_constraint=False, sparsity=True, sparsity_param=param)
    data.derivs = derivs
    all_the_weights_sparsity.append((weights, weights_0))
```

```python
all_the_weights_agi = []
for param in sample_values:
    weights_0 = optimal_weights(derivs[:1], l2_constraint=True, sparsity=False, l2_param=param)
    weights = optimal_weights(derivs, l2_constraint=True, sparsity=False, l2_param=param)
    data.derivs = derivs
    all_the_weights_agi.append((weights, weights_0))
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
for weights in all_the_weights_sparsity:
    plt.hist(weights[0])
    plt.show()
    plt.hist(weights[1])
    plt.show()
```

```python
import dill
for i, eta in enumerate(sample_values):
    pickle = {'pca': data, 'param':eta, 'weights':all_the_weights_sparsity[i],
              'notes': "Single qubit controls (pca object saved in pca key), with sparsity parameter param. Weights is a tuple, the first element is an array of weights for a sparse generator exact MQG, the second is weights for a sparse robust generator exact MQG"}
    dill.dump(pickle, open(f'sparsity{i}.pkl', 'wb'))
    pickle = {'pca': data, 'param':eta, 'weights':all_the_weights_agi[i], 'notes':"Single qubit controls (pca object saved in pca key), with agi regularization parameter param. Weights is a tuple, the first element is an array of weights for a regularized generator exact MQG, the second is weights for a regularized robust generator exact MQG"}
    dill.dump(pickle, open(f'agi{i}.pkl', 'wb'))
```
